# Leitura do dataset

```python
data = pd.read_csv('stars.csv', encoding ='utf8', sep=',')
data.head()
```

### Reorganizando o dataset

```python
nomes_colunas = ["Temperature (K)", "Luminosity(L/Lo)", "Radius(R/Ro)", "Absolute magnitude(Mv)", "Star color", "Spectral Class", "Star type"]
data = data.reindex(columns=nomes_colunas)
```

```python
data.head()
```

```python
cols = ['Star color', 'Spectral Class']
data[cols] = data[cols].apply(lambda x: pd.factorize(x)[0]+1)
data.head()
```

```python
data['Luminosity(L/Lo)'] = data['Luminosity(L/Lo)']*1000000
```

```python
data['Radius(R/Ro)'] = data['Radius(R/Ro)']*10000
```
